```python
means = []
new_df2 = df[["Total Number of Accidents Per Lakh Population - 2013","Total Number of Accidents Per Lakh Population - 2014","Total Number of Accidents Per Lakh Population - 2015","Total Number of Accidents Per Lakh Population - 2016"]]
for col in new_df2.columns:
    x = df[col].mean()
    means.append(x)
years = [2013,2014,2015,2016]
means
mean_df = pd.DataFrame({"years": years,"means":means})
mean_df
```

```python
#The highest number of accident states and least number of accident states.
df3["Total_accidents"] = df3.iloc[:,1:].sum(axis=1)
new_df3 = df3.iloc[:-1]
new_df3.tail()
```

```python
max_accident_state_index = new_df3["Total_accidents"].idxmax()
min_accident_state_index = new_df3["Total_accidents"].idxmin()
max_accident_state = new_df3.iloc[max_accident_state_index,0]
min_accident_state = new_df3.iloc[min_accident_state_index,0]
print(f'The maximum accident state is {max_accident_state} and the minium accident state is {min_accident_state}')
```
